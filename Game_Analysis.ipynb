```python
import pandas as pd 
purchase_data_df = "data/purchase_data.csv" #Purchasing file location
purchase_data_df = pd.read_csv(purchase_data_df) # Reads Purchasing File and store into Pandas data frame
purchase_data_df #displays purchase data dataframe
```

## Player Count

* Displays the total number of players


```python
#creates dataframe for TOTAL NUMBER OF INDIVIDUAL PLAYERS
tot_players_df=pd.DataFrame({
    "Total Players":[purchase_data_df['SN'].nunique()]
})
tot_players_df #display total players dataframe
```

## Purchasing Analysis (Total)

```python
#create dataframe for purchasing analysis
#total count = number of unique items
#average price per player = mean of price
#number of purchases = count of purchase id
#total revenue = sum of price
purchasing_analysis_total_df=pd.DataFrame({
    "Total Count":[purchase_data_df['Item Name'].nunique()],
    "Average Price Per Player":[purchase_data_df['Price'].mean()],
    "Number of Purchases":[purchase_data_df['Purchase ID'].count()],
    "Total Revenue":[purchase_data_df['Price'].sum()]
})
purchasing_analysis_total_df['Average Price Per Player']=purchasing_analysis_total_df['Average Price Per Player'].map("${:.2f}".format) #formats column as currency
purchasing_analysis_total_df['Total Revenue']=purchasing_analysis_total_df['Total Revenue'].map("${:.2f}".format) #formats column as currency

purchasing_analysis_total_df #displays dataframe
```

## Gender Demographics

```python
gender_group_gb=purchase_data_df.groupby(["Gender"]) #creates groupby object for GENDER
gender_count_ser=gender_group_gb['SN'].nunique() #counts number of unique players based on gender
gender_perc_players_ser=gender_count_ser/tot_players_df['Total Players'][0] #calculates percent of players for each gender

#creates dataframe for gender count and percentages
gender_dem_df=pd.DataFrame({"Total Count":gender_count_ser,
                             "Percentage of Players":gender_perc_players_ser    
})

gender_dem_df['Percentage of Players']=gender_dem_df['Percentage of Players']*100 #formats column as percentage
gender_dem_df['Percentage of Players']=gender_dem_df['Percentage of Players'].map("{:.2f}%".format) #formats column as percentage
gender_dem_df.rename_axis(None,inplace=True) #removes index title for Gender
gender_dem_df #displays dataframe
```


## Purchasing Analysis (Gender)

```python
num_of_players_gender=gender_group_gb['SN'].nunique() #counts number of unique players based on gender
purchase_count_gender_ser=gender_group_gb['Purchase ID'].count() #counts number of purchases for each gender
tot_purchase_val_gender_ser=gender_group_gb['Price'].sum() #sums all purchases for each gender
avg_purchase_price_gender_ser=tot_purchase_val_gender_ser/purchase_count_gender_ser #averages purchase price for each person/gender based on number of purchases
avg_purchase_price_per_person_gender_ser=tot_purchase_val_gender_ser/num_of_players_gender #averages purchase price for each person/gender based on number of individual players

#creates dataframe
purchasing_analysis_gender_df=pd.DataFrame({"Purchase Count":purchase_count_gender_ser,
                            "Average Purchase Price":avg_purchase_price_gender_ser,
                            "Total Purchase Value":tot_purchase_val_gender_ser,
                            "Avg Total Purchase Per Person":avg_purchase_price_per_person_gender_ser
})

#formats dataframe
purchasing_analysis_gender_df['Average Purchase Price']=purchasing_analysis_gender_df['Average Purchase Price'].map("${:.2f}".format)
purchasing_analysis_gender_df['Total Purchase Value']=purchasing_analysis_gender_df['Total Purchase Value'].map("${:.2f}".format)
purchasing_analysis_gender_df['Avg Total Purchase Per Person']=purchasing_analysis_gender_df['Avg Total Purchase Per Person'].map("${:.2f}".format)

purchasing_analysis_gender_df.rename_axis(None,inplace=True) #removes index title for Gender
purchasing_analysis_gender_df #displays dataframe
```

## Age Demographics

```python
temp_group=purchase_data_df.groupby('SN') #creates groupby object based on screen names
temp_group=temp_group['Age'].mean() #attains age of each player by taking average of their age

bins=[0,9,14,19,24,29,34,39,150] #creates bin ranges
bin_names=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"] #creates bin labels
tot_count_age=pd.cut(temp_group, bins, labels=bin_names, include_lowest=True) #bins the ages for each player

tot_count=tot_count_age.value_counts() #counts number people in each bin
perc_of_players=tot_count/tot_players_df['Total Players'][0] #percentages of players based on age

#creates dataframe
age_summary_df=pd.DataFrame({"Total Count":tot_count,
                             "Percent of Players":perc_of_players
 })

#formats cells
age_summary_df['Percent of Players']=age_summary_df["Percent of Players"]*100
age_summary_df['Percent of Players']=age_summary_df["Percent of Players"].map("{:.2f}%".format)
age_summary_df.sort_index()
```

## Purchasing Analysis (Age)

```python
bins=[0,9,14,19,24,29,34,39,150] #creates bin ranges
bin_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"] #creates bin labels
categorized_age_ser=pd.cut(purchase_data_df['Age'], bins, labels=bin_labels, include_lowest=True) #bins by age

purchase_data_df['Categorized Age']=categorized_age_ser #creates column for binned ages

ages_group_gb=purchase_data_df.groupby('Categorized Age') #groups by categorized ages
purchase_count_age_ser=ages_group_gb['Purchase ID'].count() #counts of purchased based on binned ages
purchase_count_age_ser

avg_purchase_price_age_ser=ages_group_gb['Price'].mean() #averages the prices based on binned ages
avg_purchase_price_age_ser

total_purchase_val_age_ser=ages_group_gb['Price'].sum() #sums of prices based on binned ages
total_purchase_val_age_ser

num_of_players_age=ages_group_gb['SN'].nunique() #diplays number of individual players for each age group
num_of_players_age

avg_total_purchase_per_person_age_ser=ages_group_gb['Price'].sum()/num_of_players_age #averages the purchase price based on individual players
avg_total_purchase_per_person_age_ser

#creates dataframe
purchase_analysis_age_df=pd.DataFrame({"Purchase Count":purchase_count_age_ser,
                                       "Average Purchase Price":avg_purchase_price_age_ser,
                                       "Total Purchase Value":total_purchase_val_age_ser,
                                       "Avg Total Purchase per Person":avg_total_purchase_per_person_age_ser
 })

#formats dataframe
purchase_analysis_age_df['Average Purchase Price']=purchase_analysis_age_df["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis_age_df['Total Purchase Value']=purchase_analysis_age_df["Total Purchase Value"].map("${:.2f}".format)
purchase_analysis_age_df['Avg Total Purchase per Person']=purchase_analysis_age_df["Avg Total Purchase per Person"].map("${:.2f}".format)

purchase_analysis_age_df.rename_axis(None,inplace=True) #removes index title for categorized ages
purchase_analysis_age_df
```

## Top Spenders

```python
sn_group_gb=purchase_data_df.groupby('SN') #creates groupby object based on screen name

purchase_count_sn_ser=sn_group_gb['Purchase ID'].count() #counts transactions for each player
purchase_count_sn_ser

avg_purchase_price_sn_ser=sn_group_gb['Price'].mean() #averages price paid for each player
avg_purchase_price_sn_ser

total_purchase_value_sn_ser=sn_group_gb['Price'].sum() #totals of all transactions per player
total_purchase_value_sn_ser

#creates dataframe
top_spenders_df=pd.DataFrame({"Purchase Count":purchase_count_sn_ser,
                                       "Average Purchase Price":avg_purchase_price_sn_ser,
                                       "Total Purchase Value":total_purchase_value_sn_ser
 })

#formats dataframe
top_spenders_df['Average Purchase Price']=top_spenders_df["Average Purchase Price"].map("${:.2f}".format)
top_spenders_df['Total Purchase Value']=top_spenders_df["Total Purchase Value"].map("${:.2f}".format)
top_spenders_df=top_spenders_df.sort_values(by=['Purchase Count'], ascending=False)
top_spenders_df.rename_axis(None,inplace=True)
top_spenders_df.head()
```

## Most Popular Items

```python
most_popular_item_raw_df=purchase_data_df[["Purchase ID",'Item ID',"Item Name","Price"]] #creates new dataframe based on original dataframe

most_popular_item_group_gb=most_popular_item_raw_df.groupby(["Item ID","Item Name"]) #groups object based on item id and the item name

purchase_count_popular_ser=most_popular_item_group_gb["Purchase ID"].count() #number count of times an item was purchased
purchase_count_popular_ser

item_price_popular_ser=most_popular_item_group_gb["Price"].mean() #average paid for each item 
item_price_popular_ser

total_purchase_value_popular_ser=most_popular_item_group_gb["Price"].sum() #total paid for each item
total_purchase_value_popular_ser

#creates dataframe
most_popular_df=pd.DataFrame({"Purchase Count":purchase_count_popular_ser,
                                       "Item Price":item_price_popular_ser,
                                       "Total Purchase Value":total_purchase_value_popular_ser
 })

#formats dataframe
most_popular_df['Item Price']=most_popular_df['Item Price'].map("${:.2f}".format)
most_popular_df['Total Purchase Value']=most_popular_df["Total Purchase Value"].map("${:.2f}".format)
most_popular_df=most_popular_df.sort_values(by=['Purchase Count'], ascending=False)
most_popular_df.head()
```

## Most Profitable Items

```python
#creates dataframe using same materials as above
most_profitable_df=pd.DataFrame({"Purchase Count":purchase_count_popular_ser,
                                       "Item Price":item_price_popular_ser,
                                       "Total Purchase Value":total_purchase_value_popular_ser
 }).sort_values(by=['Total Purchase Value'], ascending=False) #sorts dataframe

#formats dataframe
most_profitable_df['Item Price']=most_profitable_df['Item Price'].map("${:.2f}".format)
most_profitable_df['Total Purchase Value']=most_profitable_df["Total Purchase Value"].map("${:.2f}".format)

most_profitable_df
```
